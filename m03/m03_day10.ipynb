```python
# 계단함수
import numpy as np
import matplotlib.pyplot as plt

def step(x):
    result = x > 0.000001 # result에 True 또는 False 저장
    return result.astype(int) # 정수로 변환

x = np.arange(-10.0, 10.0, 0.1)
y = step(x) # 위에서 정의한 step함수 호출
plt.plot(x, y)
plt.show()
```

```python
# 시그모이드 함수
def sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))

x = np.arange(-10.0, 10.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.show()
```

```python
# ReLU 함수
def relu(x):
    return np.maximum(x, 0)

x = np.arange(-10.0, 10.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.show()
```

```python
# tanh함수
x = np.linspace(-np.pi, np.pi, 60)
y = np.tanh(x)
plt.plot(x, y)
plt.show()
```

```python
# MLP 순방향 패스
import numpy as np

# 시그모이드 함수
def actf(x):
    return 1/(1 + np.exp(-x))

# 시그모이드 함수의 미분치
def actf_deriv(x):
    return x * (1 - x) 

# 입력유닛의 개수, 은닉유닛의 개수, 출력유닛의 개수
inputs, hiddens, outputs = 2, 2, 1
learning_rate = 0.2
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
T = np.array([[0], [1], [1], [0]])
W1 = np.array([[0.10, 0.20], [0.30, 0.40]])
W2 = np.array([[0.50], [0.60]])
B1 = np.array([0.1, 0.2])
B2 = np.array([0.3])

# 순방향 전파 계산
def predict(x):
    layer0 = x # 입력을 layer0에 대입
    Z1 = np.dot(layer0, W1) + B1 # 행렬의 곱을 계산
    layer1 = actf(Z1) # 활성화 함수를 적용
    Z2 = np.dot(layer1, W2) + B2 # 행렬의 곱을 계산
    layer2 = actf(Z2) # 활성화 함수를 적용
    return layer0, layer1, layer2

def test():
    for x, y in zip(X, T):
        x = np.reshape(x, (1, -1)) # x를 2차원 행렬로 만든다, 입력 2차원
        layer0, layer1, layer2 = predict(x)
        print(x, y, layer2)
test()
```
