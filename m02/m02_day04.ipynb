```python
# find() 메서드 - 첫번째 태그만 가져옴(하나만)
from bs4 import BeautifulSoup # bs4 라이브러리 안에 있는 BeautifulSoup 클래스만 가져와라
html = '''
<div>
    <a href="https://www.naver.com">네이버</a><p>
    <a href="https://www.kakao.com">카카오</a>
</div> 
'''
soup = BeautifulSoup(html, 'html.parser')
print(soup.find('a')) # 해당 태그 전체(첫번재 태그만)
print(soup.find('a').text) # 해당 태그의 텍스트 부분(내용 부분)
print(soup.find('a').get('href')) # href 속성에 해당하는 속성값
```

```python
# find_all() - 해당 태그 모두 다 가져옴(결과는 리스트형식으로)
html ='''
<ul>
    <li id="movie">영화</li>
    <li>여행</li>
    <li>독서</li>
</ul>
'''
soup = BeautifulSoup(html, 'html.parser')
print(soup.find_all('li')) # li 태그 전체(리스트형식으로)
print(soup.find_all('li')[0]) # 0번 인덱스 값의 태그 전체
print(soup.find_all('li')[0].text) # 0번 인덱스 값의 텍스트부분만
print(soup.find_all('li')[1].text)
print(soup.find_all('li')[2].text)
print(soup.find_all('li')[0].get('id')) # id라는 속성의 값을 출력
```

```python
# select_one() 
html = '''
<div>
    <a href="https://www.naver.com">네이버</a><p>
    <a href="https://www.kakao.com">카카오</a>
</div> 
'''
soup = BeautifulSoup(html, 'html.parser')

print(soup.select_one('a')) # 첫 번째 <a> 태그 전체 출력
print(soup.select_one('a').text)
print(soup.select_one('a').get('href'))
```

```python
# select()
html ='''
<ul>
    <li id="movie">영화</li>
    <li>여행</li>
    <li>독서</li>
</ul>
'''
soup = BeautifulSoup(html, 'html.parser')

print(soup.select('li')) # <li> 태그 전체 -> 리스트로 출력
print(soup.select('li')[0])
print(soup.select('li')[1])
print(soup.select('li')[2])
print(soup.select('li')[0].text)
print(soup.select('li')[1].text)
print(soup.select('li')[2].text)
```

```python
# class 속성 - 공통적인 특징을 줄 때 사용
from bs4 import BeautifulSoup

html = '''
<div>
    <div class="gnb">뉴스</div>
    <div class="gnb">지도</div>
</div>
'''
soup = BeautifulSoup(html, 'html.parser')

# find_all(name='태그명', attrs={'속성이름':'속성값})
print(soup.find_all(name='div')) # <div> 태그 다 가져옴 -> 3개
print()
print(soup.find_all(name='div', attrs={'class' : 'gnb'})) 
print(soup.find_all(name='div', attrs={'class' : 'gnb'})[0])
print(soup.find_all(name='div', attrs={'class' : 'gnb'})[1])
print(soup.find_all(name='div', attrs={'class' : 'gnb'})[0].text)
print(soup.find_all(name='div', attrs={'class' : 'gnb'})[1].text)
```

```python
from bs4 import BeautifulSoup

html = '''
<div>
    <div class="gnb">뉴스</div>
    <div class="gnb">지도</div>
</div>
'''
soup = BeautifulSoup(html, 'html.parser')

# select()
# class -> .  
print(soup.select('div.gnb')) # class가 gnb인 <div>태그 전체 리스트로 출력
print(soup.select('div.gnb')[0])
print(soup.select('div.gnb')[1])
print(soup.select('div.gnb')[0].text)
print(soup.select('div.gnb')[1].text)
```

```python
# id 속성
from bs4 import BeautifulSoup

html = '''
<div id="container">
    <div id="left">왼쪽 영역</div>
    <div id="right">오른쪽 영역</div>
</div>
'''
soup = BeautifulSoup(html, 'html.parser')

# find()
print(soup.find(name='div', attrs={'id' : 'left'}))
print(soup.find(name='div', attrs={'id' : 'right'}))
print(soup.find(name='div', attrs={'id' : 'left'}).text)
print(soup.find(name='div', attrs={'id' : 'right'}).text)
```

```python
# 일반 웹페이지 정보 가져오기
import requests

url = 'https://www.naver.com'
response = requests.get(url) # url에 접속해서 가지고 온 결과를 response변수에 담음
print(response.text) # text속성을 지정하면 html 정보를 볼 수 있음
print('-' * 100)
print(f'응답코드 : {response.status_code}') # 응답코드 출력 -> 200대가 나오면 정상
```

```python
# 검색결과 웹페이지 정보 가져오기
import requests

url = 'https://search.naver.com/search.naver'
param = {'query' : '파이썬'} # 검색어가 파이썬(url + param)
response = requests.get(url, params=param)
print(response.text)
```

# CGV 영화 제목 추출하기

```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.cgv.co.kr/movies'
response = requests.get(url)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
# select 계열
tags = soup.select('strong.title') # strong태그의 class속성의 값 title
tags
```

```python
for i in tags:
    print(i.text)
```

```python
for idx, tag in enumerate(tags, start=1):
    print(f'{idx}위 : {tag.text}')
```

```python
# find() 계열
tags2 = soup.find_all('strong', {'class':'title'})
tags2
```

```python
for index, movie in enumerate(tags2, start=1):
    print(f'{index}위 : {movie.text}')
```
