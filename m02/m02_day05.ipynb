```python
# 다음 뉴스
import requests
from bs4 import BeautifulSoup

url = 'https://news.daum.net/tech'
response = requests.get(url)
# 한글이 깨지는 경우 인코딩 설정을 'utf-8'로 해줌
response.encoding = 'utf-8'
html = response.text
soup = BeautifulSoup(html, 'html.parser')
news_result = soup.select('strong.tit_txt')

for news in news_result:
    print(news.text)
```

```python
# Yes24 에세이 베스트셀러 가져오기 - 책이름, 작가이름, 출판사
import requests
from bs4 import BeautifulSoup

url = 'https://www.yes24.com/product/category/bestseller?pageNumber=1&pageSize=24&categoryNumber=001001047'
response = requests.get(url)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
book_list = soup.select('div.itemUnit')
# 책 제목
for title in book_list:
    print(title.select_one('a.gd_name').text)
```

```python
# 저자 <span class="authPub info_pub"
for author in book_list:
    print(author.select_one('span.authPub').text)
```

```python
# 저자 - '저'글자 제외 <span class=authPub.info_auth > 하위 <a>에 '저'빼고 저자만 저장되어있음
for author in book_list:
    print(author.select_one('span.authPub > a').text) 
```

```python
# 출판사 <span class="authPub info_pub"
for publisher in book_list:
    print(publisher.select_one('span.info_pub').text)
```

```python
# pandas DataFrame으로 내보내기
import pandas as pd

book = [] # 결과를 저장할 리스트 -> 데이터프레임이 됨
for b in book_list: # book_list : 책 전체내용
    title = b.select_one('a.gd_name').text # 책 제목
    author = b.select_one('span.authPub > a').text # 저자
    publisher = b.select_one('span.info_pub').text # 출판사
    book.append({
        '도서 제목' : title, 
        '저자' : author, 
        '출판사' : publisher
        })
df_book = pd.DataFrame(book)
df_book
```

```python
# csv파일로 내보내기
df_book.to_csv('./output/book.csv', index=False) # 인덱스번호 제외하고 저장
```

# 네이버 증권 -> 환율 정보 가져오기
# 웹페이지 안에 웹페이지 있는 형태 iframe 을 찾아서 주소창에 다시 입력

```python
import requests
from bs4 import BeautifulSoup

url = 'https://finance.naver.com/marketindex/exchangeList.naver'
response = requests.get(url)
html = response.text
soup = BeautifulSoup(html, 'html.parser')

# 나라별 -> 한 행씩 가져오기
# body > div > table > tbody > tr:nth-child(1) > td.tit > a
# nth-child(1) : 1번째 자식
rows = soup.select('tbody > tr')
rows
```

```python
# 나라 이름
for row in rows:
    title = row.select_one('td.tit').text.strip() # 양쪽 공백 제거
    print(title)
```

```python
# 환율 
for row in rows:
    sale = row.select_one('td.sale').text
    print(sale)
    print(type(sale)) # str 타입 -> 변환하기
```

```python
# 국가번호를 입력하면 -> 해당 나라의 통화를 원으로 계산해주는 환율 계산기 구현

# 국가번호  1. 미국 2. 유럽 3. 일본 4. 중국

def get_exchange_rate(menu): # menu : 국가번호
    '''특정 나라의 통화를 실수형으로 가져오는 함수'''
    url = 'https://finance.naver.com/marketindex/exchangeList.naver'
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    exchange_rates = soup.select('tbody > tr > td.sale') # 환율 : 문자열 -> 실수형으로 형변환해주기!
    exchange_rate = float(exchange_rates[menu-1].text.replace(',', ''))  # 실수형으로 변환, 천단위(,)구분자 없애기 -> 실제 나라(메뉴)에 해당하는 환율이 exchange_rate에 저장
    return exchange_rate    
```

```python
menu = int(input('''원하시는 국가를 선택해주세요!
                    ==========================
                    1. 미국
                    2. 유럽
                    3. 일본
                    4. 중국
                    =========================='''))
unit = ['달러', '유로', '엔', '위안']
money = int(input(f'환율 계산할 금액을 입력해주세요. (단위 : {unit[menu-1]}) : '))

# 일본은 100엔을 기준으로 하기 때문에 따로 계산해주기
if menu == 3:
    result_money = get_exchange_rate(menu) * (money // 100)
else:
    result_money = get_exchange_rate(menu) * money

print(f'{money}{unit[menu-1]}은(는) 원으로 환전하면 {result_money:.1f}원입니다.')
```
